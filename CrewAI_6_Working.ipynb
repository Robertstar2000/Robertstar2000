```python
import os
import crewai
#from langchain_openai import ChatOpenAI
from crewai import Crew, Process, Agent, Task
# Importing crewAI tools
from crewai_tools import DirectorySearchTool, FileReadTool, SerperDevTool, WebsiteSearchTool
from crewai import Process
client = Groq(
    api_key=os.environ.get("GROQ_API_KEY"),
)

# Instantiate tools
docs_tool = DirectorySearchTool
file_tool = FileReadTool()
search_tool = SerperDevTool()
web_rag_tool = WebsiteSearchTool()

#goal_4 = f'Follow the plan and ask the writer to respond to each section {goal_2}'
# print (goal_3)


# Create agents
Conductor = Agent(
    role= goal_3,
    goal=goal_3,
    backstory='An expert conductor of tasks uses goals to decide what task to do next and evaluates how many repetitions of each task are needed to complete each task then deligates to the best agent and moves on to the next task.',
    allow_delegation='True',
    # llm='Meta-Llama-3-8B-Instruct',
    cache=True,
    #model='Meta-Llama-3-8B-Instruct',
    verbose=True,
    max_iter=55,
    max_rpm=2
)

researcher = Agent(
    role=' Research Analyst',
    goal='Provide up-to-date data on the current task or subtask',
    backstory='An expert analyst with a keen eye',
    #tools=[search_tool, web_rag_tool],
    cache=True,
    allow_delegation=True,
    verbose=True,
    #llm='Meta-Llama-3-8B-Instruct',
    #model='Meta-Llama-3-8B-Instruct',
    max_iter=60,
    max_rpm=2,
)

builder = Agent(
    role='Content Creator',
    goal='Craft engaging content using knowedge and instructions from the conductor',
    backstory='A skilled writer with passion.',
    allow_delegation=True,
    tools=[search_tool, web_rag_tool],
    #llm='Meta-Llama-3-8B-Instruct',
    #model='Meta-Llama-3-8B-Instruct',
    cache=True,
    verbose=True,
    max_iter=25,
    max_rpm=2,
)

# Define tasks

Sequencer = Task(
    description='Step by step reaseasrch eack outline item build it and then create a output .',
    expected_output='Direction to the other Tasks and agents',
    agent=Conductor,  # Change the agent to the crew object
    allow_delegation='True',
)
research = Task(
    description='Research the task or subtask provided by the conductor and provide a summary.',
    expected_output='A reasearched summary in list format with five items.',
    agent=researcher,  # Change the agent to the crew object
    allow_delegation='True',
)

Output = Task(
    description=(
    """For tasks output a header <Title> followed by the <full expanded content
    -For subtasks output the <full expanded content (without a bold title)
    -To create content describe, expain and provide context to understand
    -Use profesional but human sounding writing style
    -Ues the writing style provided
     print(full_expanded_content). Use
    knowledge from telemetry and privious outputs
    """),
    expected_output=(
    """
    1 Print A OUTLINE of the tasks and subtasks with number Example <3.6>
    2-The resulting content is appended into a single file called Output
    """
    ),
    agent=builder,  # Change the agent to the crew object
    allow_delegation='True')

# Assemble,  a crew
crew = Crew(
  agent=[Conductor, researcher, builder],  # Change Task to Agent
  tasks=[research, Sequencer, Output],
  memory=True,
  verbose=True,
  max_iter=50,
  max_rpm=2,
)

# Execute tasks
crew.kickoff()
```
